Soybean Production by Top 5 States and Yield Efficiency

```python
#Step 1: Clean your data
df_clean = df.copy()
df_clean = df_clean.dropna(subset=[
    'SOYABEAN PRODUCTION (1000 tons)',
    'SOYABEAN AREA (1000 ha)'])
df_clean = df_clean[
    (df_clean['SOYABEAN PRODUCTION (1000 tons)'] > 0) &
    (df_clean['SOYABEAN AREA (1000 ha)'] > 0)]


df_clean['Yield Efficiency']=df_clean['SOYABEAN PRODUCTION (1000 tons)']/df['SOYABEAN AREA (1000 ha)']
top_soybean = df_clean.groupby('State Name').agg({
    'SOYABEAN PRODUCTION (1000 tons)': 'sum',
    'SOYABEAN AREA (1000 ha)': 'sum'
 }).reset_index()

top_soybean['Yield Efficiency'] = (top_soybean['SOYABEAN PRODUCTION (1000 tons)'] / top_soybean['SOYABEAN AREA (1000 ha)']).round(2)

top_soybean = top_soybean.nlargest(5, 'SOYABEAN PRODUCTION (1000 tons)')

fig = px.pie(
    top_soybean ,
    names='State Name',
    values='SOYABEAN PRODUCTION (1000 tons)',
    title='Top 5 States for Soybean Production',
    hole=0.3,
    hover_data=['Yield Efficiency']
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Insights: Soybean Production Share by Top 5 States
**Madhya Pradesh Leads National Output**
With 49.8% of total soybean production, Madhya Pradesh is the dominant contributor.
**Maharashtra Is a Strong Second**
Contributing 26.6%, Maharashtra plays a key role in national supply.
**High Regional Concentration**
The top 3 states (MP, Maharashtra, Rajasthan) account for over 85% of total production.
Indicates a strong geographic concentration in central India, guiding targeted policy and investment.

Impact of Area Cultivated on Production (Rice, Wheat, Maize)

```python
data_area_production = df.groupby("Year")[[
    "RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)",
    "WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)",
    "MAIZE PRODUCTION (1000 tons)", "MAIZE AREA (1000 ha)"
]].sum().reset_index()
data_area_production 
```

```python
rice = data_area_production[["RICE AREA (1000 ha)", "RICE PRODUCTION (1000 tons)"]].rename(
    columns={"RICE AREA (1000 ha)": "Area", "RICE PRODUCTION (1000 tons)": "Production"}
)
rice["Crop"] = "Rice"

wheat = data_area_production[["WHEAT AREA (1000 ha)", "WHEAT PRODUCTION (1000 tons)"]].rename(
    columns={"WHEAT AREA (1000 ha)": "Area", "WHEAT PRODUCTION (1000 tons)": "Production"}
)
wheat["Crop"] = "Wheat"

maize = data_area_production[["MAIZE AREA (1000 ha)", "MAIZE PRODUCTION (1000 tons)"]].rename(
    columns={"MAIZE AREA (1000 ha)": "Area", "MAIZE PRODUCTION (1000 tons)": "Production"}
)
maize["Crop"] = "Maize"

# Combine all crops
combined = pd.concat([rice, wheat, maize])

# Create interactive scatter plot with regression lines
fig = px.scatter(
    combined,
    x="Area",
    y="Production",
    color="Crop",
    labels={"Area": "Area Cultivated (1000 ha)", "Production": "Production (1000 tons)"},
    title="Correlation between Area and Production",
    hover_data=["Crop", "Area", "Production"]
)

fig.show()
```

The chart shows how much crop is produced based on how much land is used. If the dots go upward as you move right, it means more land leads to more production â€” a positive correlation.

Insights for Area Cultivated on Production (Rice, Wheat, Maize)
Rice :

Shows a strong positive correlation â€” as area cultivated increases, production rises consistently.

Clustered in higher area and production ranges, suggesting rice is grown on larger plots and yields high output.

Wheat :

Also positively correlated, but with slightly more spread than rice.

Appears to be cultivated on moderately sized areas with decent production, but not as high-yielding as rice per hectare.

Maize :

Displays a positive trend, though with more variability.

Some points show high production even with smaller cultivated areas, hinting at higher efficiency  in certain cases.

Maize is cultivated on smaller areas but can still achieve competitive production, indicating potential for efficient land use or high-yield varieties

Rice vs. Wheat Yield Across States

```python
data_yield = df.groupby("State Name")[[
    "RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)",
    "WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)"
]].sum().reset_index()


data_yield["Rice Yield"] = data_yield["RICE PRODUCTION (1000 tons)"] / data_yield["RICE AREA (1000 ha)"]
data_yield["Wheat Yield"] = data_yield["WHEAT PRODUCTION (1000 tons)"] / data_yield["WHEAT AREA (1000 ha)"]

data_yield = data_yield[["State Name", "Rice Yield", "Wheat Yield"]]
data_yield
```

```python
plt.figure(figsize=(12, 6))

# Melt the dataframe to long format for easy plotting
data_melted = data_yield.melt(id_vars="State Name", #values need to be fixed
                              value_vars=["Rice Yield", "Wheat Yield"], #cols in rows
                              var_name="Crop", 
                              value_name="Yield")

# Create  grouped bar chart
fig = px.bar(data_melted, 
             x="State Name", 
             y="Yield", 
             color="Crop", 
             barmode="group",
             title="Rice vs Wheat Yield Across States",
             labels={"Yield": "Yield (tons per hectare)"})

# Show chart
fig.show()
```

Insights 
**Top Performers**
Punjab excels in both rice and wheat yield.

Andhra Pradesh, Telangana, Kerala lead in rice yield.

Haryana, Uttar Pradesh, Madhya Pradesh shine in wheat yield.

**Crop Dominance**
Southern states favor rice; wheat yield is low.

Northern states favor wheat; rice yield is moderate.

Northeast and Goa/Sikkim show low yields for both crops.

**Action Points**
Boost support in low-yield states.

Use Punjab as a benchmark for best practices.

Consider crop diversification where one crop underperforms.
